```python
from selenium.webdriver.common.by import By
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301')
```

```python
products = driver.find_elements(By.XPATH,"//span[@class='a-size-base-plus a-color-base a-text-normal']")
prices = driver.find_elements("xpath","//span[@class='a-price-whole']")

for i in products:
    print(i.text)
    
for i in prices:
    print(i.text)
```

```python
print (len(products))
print (len(prices))
```

```python
import pandas as pd
pd.DataFrame({'products':products,'prices':prices})
```

```python
import urllib
df=pd.DataFrame([],columns=['a','b','c','d','e'])
products_Sections = driver.find_elements(By.XPATH,"//div[@class='a-section a-spacing-base']")
for section in products_Sections:
    try:
        title=section.find_element(By.XPATH,".//span[@class='a-size-base-plus a-color-base a-text-normal']")
        print(title.text)
        price=section.find_element(By.XPATH,".//span[@class='a-price-whole']")
        print(price.text)
        image_src=section.find_element(By.XPATH,".//img[@class='s-image']")
        url=image_src.get_attribute('src')
        print(url)
        rating=section.find_element(By.XPATH,".//div[@class='a-row a-size-small']/span").get_attribute('aria-label')
        print(rating.split()[0])
        total_rating=section.find_element(By.XPATH,".//span[@class='a-size-base s-underline-text']")
        print(total_rating.text)
        urllib.request.urlretrieve(url,"{}.jpg".format(title.text))
        df.loc[len(df)]=[title.text,price.text,url,total_rating.text,rating.split()[0]]
    except:
        print('error')
```
